```python
categorical_features=["gender","ethnicity","education_level","income_level","smoking_status","employment_status"]
numerical_features=["age","alcohol_consumption_per_week","physical_activity_minutes_per_week","diet_score","sleep_hours_per_day","screen_time_hours_per_day","bmi","waist_to_hip_ratio","systolic_bp","diastolic_bp","heart_rate","cholesterol_total","hdl_cholesterol","ldl_cholesterol","triglycerides","family_history_diabetes","hypertension_history","cardiovascular_history"]
```

```python
DF_train_DS.head()
```

```python
DF_train_DS_X=DF_train_DS.drop(['diagnosed_diabetes'],axis=1)
DF_train_DS_Y=DF_train_DS['diagnosed_diabetes']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.impute import SimpleImputer

preprocessor=ColumnTransformer([
    ('cat',OneHotEncoder(),categorical_features),
    ('num', StandardScaler(), numerical_features),

]
    
)


transformed_train_x=preprocessor.fit_transform(DF_train_DS)
transformed_test_x=preprocessor.transform(DF_test_DS)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(transformed_train_x,DF_train_DS_Y,test_size=0.2,random_state=True)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

modelXG = xgb.XGBRegressor(objective='reg:squarederror',
                         n_estimators=200, random_state=42)
modelXG.fit(X_train, y_train)
y_pred = modelXG.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f'RMSE: {rmse:.3f}')
print(f'R²: {r2:.3f}')
```
