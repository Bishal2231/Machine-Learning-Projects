```python
df=pd.read_csv('/kaggle/input/youtube-top-100-songs-2025-dataset/youtube-top-100-songs-2025.csv')
df.head()
```

```python
df.shape
```

```python
# Group by 'categories' and sum view_count
category_views = df.groupby('categories')['view_count'].sum().reset_index()
```

```python
category_views
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='categories', y='view_count', data=category_views)
plt.title("Total View Count by Category")
plt.ylabel("Total Views")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.show()
```
